```python
import requests
import PyPDF2
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import random

# Load pre-trained model and tokenizer
model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Set maximum length for generated MCQs
max_new_tokens = 200  # Increase the max_new_tokens value

# Define the URL of the PDF file
pdf_url = "https://github.com/akaiketech/internship-assignment-nlp/raw/main/Dataset/chapter-2.pdf"  # i chose chapter -2  only ,other take more time its  a big data

# Download the PDF file
response = requests.get(pdf_url)
pdf_content = response.content

# Load the PDF file
with open("chapter-2.pdf", "wb") as file:
    file.write(pdf_content)

#  the downloaded PDF file
with open("chapter-2.pdf", "rb") as file:
    pdf_reader = PyPDF2.PdfReader(file)

    # Extract the text from the PDF
    chapter_text = ""
    for page in pdf_reader.pages:
        chapter_text += page.extract_text()

# Split the chapter text into smaller chunks
chunk_size = 1000  # Adjust the chunk size as needed
chunks = [chapter_text[i:i + chunk_size] for i in range(0, len(chapter_text), chunk_size)]

# Define question templates and possible answer choices
question_templates = [
    "What is the main idea of the passage?",
    "Which statement best describes <blank>?",
    "What is the significance of <blank>?",
    "Which of the following is an example of <blank>?",
    "According to the passage, what is the role of <blank>?",
    "What can be inferred about <blank> from the passage?",
]

possible_answer_choices = [
    ["Option A", "Option B", "Option C", "Option D"],
    ["Option 1", "Option 2", "Option 3", "Option 4"],
    ["Option X", "Option Y", "Option Z"],
]

# Generate MCQs for each chunk
all_generated_mcqs = []
for chunk in chunks:
    # Tokenize the chunk
    input_ids = tokenizer.encode(chunk, add_special_tokens=True, truncation=True, max_length=max_new_tokens, return_tensors="pt")

    # Generate MCQs
    output = model.generate(
        input_ids,
        num_return_sequences=5,
        num_beams=5,
        pad_token_id=model.config.eos_token_id,
        max_new_tokens=max_new_tokens
    )

    # Decode the generated MCQs
    generated_mcqs = []
    for generated in output:
        question_template = random.choice(question_templates)
        answer_choices = random.choice(possible_answer_choices)
        
        question = question_template.replace("<blank>", tokenizer.decode(generated, skip_special_tokens=True))
        mcq = {
            "question": question,
            "answer_choices": answer_choices
        }
        generated_mcqs.append(mcq)
    
    all_generated_mcqs.extend(generated_mcqs)

# Print all the generated MCQs
for i, mcq in enumerate(all_generated_mcqs):
    print(f"MCQ {i+1}: {mcq['question']}")
    print("Answer Choices:")
    for j, choice in enumerate(mcq['answer_choices']):
        print(f"{j+1}. {choice}")
    print()
```
